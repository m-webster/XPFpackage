# Chapter 5 Examples: Weighted Graph and Hypergraph States

This notebook illustrates the algorithms for representing Weighted Graph and Hypergraph States as XP codese

```python
import add_parent_dir
import numpy as np
from common import *
from NSpace import *
from XPAlgebra import *
from XPCodes import *
from graph_states import *

## Example from XS paper
edges = [(1,2,3)]
weights = None

## Hypergraph state - Union Jack 
# edges = [(0,1,3),(0,2,3),(0,1),(1,3),(3,2),(2,0)]
# weights = None

## Weighted graph state - universal MBQC
# edges = [(0,1),(2,3),(0,3),(1,2)]
# weights = [2,2,1,1]

GState = graphState(edges,weights)
setVerbose(True)    
G = GState.XPCode()
N = GState.N
setVerbose(False)
print('\n')
print('Stabilizer Generators of Graph State: G =')
print(XP2Str(G,N))
print('\n')
S = GState.State()
print('Graph State: S=')
print(State2Str(S,N) )
print('\n')

C = Code(G,N)
SCheck = getVal(C,'Codewords')[0]
print('Checking Generators: ',StateEqual(SCheck,S))
```
