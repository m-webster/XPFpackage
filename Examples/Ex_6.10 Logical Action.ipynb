# Examples 6.10, 6.11: Logical Actions

This notebook demonstrates how to determine which diagonal logical actions arise in an arbitrary XP Code.

- Example 6.10 looks at this for Code 1
- Example 6.11 looks at Code 2 - un-comment in the code where indicated to run this example

```python
import add_parent_dir
import numpy as np
from common import *
from NSpace import *
from XPAlgebra import *
from XPCodes import *

## Code 1
genstr = 'XP_8(8|0000000|6554444),XP_8(7|1111111|1241234),XP_8(1|1110000|3134444)'

## Code 2
## Un-comment to use Code 2
genstr = 'XP8(0|0000000|1322224)\nXP8(12|1111111|1234567)'

setVerbose(False)
G, N = str2XP(genstr)
n = XPn(G)
print('The Code Generators are: G =')
print(XP2Str(G,N),"\n")
C = Code(G,N)
LO = getVal(C,'LO')
LX,LZ = splitDiag(LO)
print('The diagonal logical generators are: LZ =')
for A in LZ:
    print(XP2Str(A,N),'f-vector',C.Fvector(A))
wI = getVal(C,'wI')
print('\n')
print('Append wI = ',XP2Str(wI,N))
LZ = np.vstack([[wI],LZ])
print('The f-vectors corresponding to the elements of wI + LZ are: FZ =')
FZ = ZMat([C.Fvector(A) for A in LZ])
print(ZmatPrint(FZ,2*N))
LD,FD = getVals(C,['LD','FD'])
print('\n')
print('The Howell Basis of FZ is: FD =')
print(ZmatPrint(FD,2*N))
print('Corresponding logical operators')
for A in LD:
    print(XP2Str(A,N),'f-vector',C.Fvector(A))

```
